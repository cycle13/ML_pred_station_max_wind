```python
# set up things
%matplotlib inline
%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings('ignore')

# load libraries
import numpy as np
import pandas as pd
import xarray as xr

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

import metpy.calc as mpcalc
from metpy.plots import StationPlot, sky_cover
from metpy.units import units

from nmc_met_io.read_micaps import read_micaps_4
from nmc_met_graphics.plot.china_map import add_china_map_2cartopy
```

```python
filename = "20062520.003"
data = read_micaps_4(filename)
data.info()
```

```python
# 绘制图像
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection=ccrs.LambertConformal(central_longitude=120))
data.plot(ax=ax, transform=ccrs.PlateCarree(), cbar_kwargs={'shrink': 0.8})
ax.coastlines()
ax.gridlines()
ax.set_extent([80,130,15,54], crs=ccrs.PlateCarree())
```
